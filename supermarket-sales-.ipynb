```python
df['Profit'] = df['Total'] - df['cogs']
```

```python
# Convert 'Date' column to datetime safely (auto-detect format)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')  # Converts invalid to NaT

# Optional: Check if any date conversion failed
print("Missing dates:", df['Date'].isnull().sum())

# Extract Month and Day if dates are successfully parsed
df['Month'] = df['Date'].dt.month_name()
df['Day'] = df['Date'].dt.day_name()



# Show result
print(df[['Date', 'Month', 'Day']].head())
```

```python
df.groupby('Month')['Total'].sum().plot(kind='bar')
```

```python
top_city = df.groupby('City')['Total'].sum().sort_values(ascending=False)
print(top_city)
```

```python
top_product = df.groupby('Product line')['Total'].sum().sort_values(ascending=False)
print(top_product)
```

```python
top_gender = df.groupby('Gender')['Total'].sum().sort_values(ascending=False)
print(top_gender)
```

```python
df['Date'] = pd.to_datetime(df['Date'])  # Safe conversion
df['Month'] = df['Date'].dt.month_name()

monthly_sales = df.groupby('Month')['Total'].sum().sort_values(ascending=False)
print(monthly_sales)
```

```python
df['Day'] = df['Date'].dt.day_name()
top_day = df.groupby('Day')['Total'].sum().sort_values(ascending=False)
print(top_day)
```

```python
payment = df.groupby('Payment')['Total'].sum().sort_values(ascending=False)
print(payment)
```
